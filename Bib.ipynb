## Regula Falsi (automatically converge)

```python
def regula_falsi(f,x0,x1,tol):
    global solution
    
    if abs(f(x0))<=tol:
        return print(f'The solution is x={x0} and f(x)={f(x0)}\n')
    elif abs(f(x1))<=tol:
        return print(f'The solution is x={x1} and f(x)={f(x1)}\n')

    else:
        f0 = f(x0)
        f1 = f(x1)
        a = (f1-f0)/(x1-x0)
        b = f0 - a*x0
        x_new = -b/a
        if f(x_new)>0 and f0>0:
            iteration_and_other(f,x_new,solution)
            return regula_falsi(f,x_new,x1,tol)
        elif f(x_new)>0 and f1>0:
            iteration_and_other(f,x_new,solution)
            return regula_falsi(f,x0,x_new,tol)
        elif f(x_new)<0 and f1<0:
            iteration_and_other(f,x_new,solution)
            return regula_falsi(f,x0,x_new,tol)
        elif f(x_new)<0 and f0<0:
            iteration_and_other(f,x_new,solution)
            return regula_falsi(f,x_new,x1,tol)
        elif f(x_new)==0:
            iteration_and_other(f,x_new,solution)
            return print(f'The solution is x={x_new} and f(x)={f(x_new)}\n')
```

## Secant

```python
def secant_method(f,x0,x1,tol):
    global solution
    f0 = f(x0)
    f1 = f(x1)

    iteration_and_other(f,x1,0)
    
    if abs(f0)<=tol:
        return print(f'The solution is x={x0} and f(x)={f0}\n')
    elif abs(f1)<=tol:
        return print(f'The solution is x={x1} and f(x)={f1}\n')

    else:
        fprime = (f1-f0)/(x1-x0)
        x_new = x1 - f1/fprime
        return secant_method(f,x0,x_new,tol)
```

## Romberg

```python
def romberg(f, a, b, max_iter):
    """Calculate the integral from the Romberg method.

    Args:
        f (function): the equation f(x).
        a (float): the initial point.
        b (float): the final point.
        max_iter (int).

    Returns:
        xi (float): numerical approximation of the definite integral.
    """
    # Initialize the Romberg integration table
    r = np.zeros((max_iter, max_iter))

    # Compute the trapezoid rule for the first column (h = b - a)
    h = b - a
    r[0, 0] = 0.5 * h * (f(a) + f(b))

    for i in range(1, max_iter):
        h = 0.5 * h  # Halve the step size
        # Compute the composite trapezoid rule
        sum_f = 0
        for j in range(1, 2**i, 2):
            x = a + j * h
            sum_f += f(x)
        r[i, 0] = 0.5 * r[i - 1, 0] + h * sum_f

        # Richardson extrapolation for higher order approximations
        for k in range(1, i + 1):
            r[i, k] = r[i, k - 1] + \
                (r[i, k - 1] - r[i - 1, k - 1]) / ((4**k) - 1)

    return r

a =  #beginning
b =  #end
it =  #iterations
sol = romberg(the_func,a,b,it)
```

## Fast Fourier Transform (fft)

```python
plt.rcParams['figure.figsize'] = [12, 12]
plt.rcParams.update({'font.size': 18})

nu = 6 # param constant
L = 20     # Length of domain
N = 100   # Number of discretization points
dx = L/N
x = np.arange(-L/2,L/2,dx) # Define x domain

# Define discrete wavenumbers or else
w = 2*np.pi*np.fft.fftfreq(N, d=dx)

# Initial condition
u0 = np.cos(np.pi*x/10)

# Simulate PDE in spatial domain
dt = 0.025
t = np.arange(0,100*dt,dt)

def rhsBurgers(u,t,w,nu): # change in function of obtain differential equations
    uhat = np.fft.fft(u)
    d_uhat = (1j)*w*uhat
    ddd_uhat = -(1j)*np.power(w,3)*uhat
    d_u = np.fft.ifft(d_uhat)
    ddd_u = np.fft.ifft(ddd_uhat)
    du_dt = nu * u * d_u - ddd_u
    return du_dt.real

u = odeint(rhsBurgers,u0,t,args=(w,nu))

# Waterfall plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# change the label and labelpad for position
ax.set_xlabel('x',labelpad=15)
ax.set_ylabel('Time',labelpad=15)
ax.set_zlabel('u(x, t)',labelpad=-30)

u_plot = u[0:-1:10,:]
for j in range(u_plot.shape[0]):
    ys = j*np.ones(u_plot.shape[1])
    ax.plot(x,ys,u_plot[j,:])
    
# Image plot
plt.figure(figsize=(12, 6))
im = plt.imshow(u, aspect=8, extent=[x.min(), x.max(), t[0], t[-1]], cmap='viridis')
plt.colorbar(im, label='u(x, t)')
plt.xlabel('x')
plt.ylabel('Time')
plt.show()
```
